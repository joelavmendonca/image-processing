Crie uma função que tenha como entradas:
- Inteiro **m** que corresponde às linhas de uma matriz;
- Inteiro **n** que corresponde às colunas de uma matriz;

e que retorne:
- Uma matriz **X** de shape (m, n), em que cada elemento X[i,j] seja a distância Euclidiana entre a posição [i,j] e a posição [0,0].

Dica: criar novas matrizes com as funções [*np.meshgrid*](https://github.com/lrittner/Computacao-Visual/blob/master/tutoriais/02_tutorial_numpy_1_6.ipynb) e [*np.arrange*](https://github.com/lrittner/Computacao-Visual/blob/master/tutoriais/02_tutorial_numpy_1_1.ipynb) para realizar as operações e obter os valores que comporão a matriz X.

Teste a função com os valores **m=5** e **n=4**.

```python
#def fun(m, n):
import math
import numpy as np
x0 = 0 # posição '0' referente a linha
y0 = 0 # posição '0' referente a coluna

def fun(m,n):
  linhas, colunas = np.meshgrid(np.arange(m), np.arange(n), indexing='ij') # cria as arrays de linhas e colunas, onde cada elemento representa os indices de linhas e colunas respectivamente
  X = np.sqrt((linhas - x0)**2 + (colunas - y0)**2) # calculo da distancia euclidiana entre cada elemento [i,j] com a posição [0,0]
  return X

print(fun(5,4))

```

**c)** Círculo de raio 40, com centro em (0,0);

```python
import numpy as np
import matplotlib.pyplot as plt

raio = 40

x = np.linspace(-raio*2, raio*2, 500) # retorna 1000 números em x igualmente espaçados que começa em -80 e termina em 80
y = np.linspace(-raio*2, raio*2, 500) # retorna 1000 números em y igualmente espaçados que começa em -80 e termina em 80
X, Y = np.meshgrid(x,y) # cria uma grade quadrangular a partir das arrays unidimensional x e y
F = (X)**2 + (Y)**2-raio**2 # realiza o calculo do circulo

circulo = F <= 0 # cria uma array binaria do circulo F; retonando uma matriz onde "True" significa que são regiões onde F<=0 e "False" para regioes onde F>0

plt.figure(figsize=(8, 6))
plt.imshow(circulo, cmap='binary', extent=[-raio*2, raio*2, -raio*2, raio*2])
plt.show()
```

## Parte 5 - Geração automática de imagens sintéticas e manipulação de arquivos (opicional para EA979)

### Exercício 5.1

**a)** Quando geramos imagens sintéticas, estamos interessados em criar um conjunto de imagens com algumas características pré-definidas, mas com certa variabilidade. Para exemplificar esse processo, vamos automatizar a criação dos quadrados do exercício 3.1.

Crie uma função que tenha como entradas:
- Inteiro **n** que corresponde ao número de imagens a serem criadas;
- Inteiro **h** que corresponde à "altura" de cada imagem (com "largura" 2\***h**);

e que retorne:
- Um array **I** de shape (n, h, 2\*h, 3), em que cada **I**[n] corresponda a uma imagem similar à imagem gerada no exercício 3.1.b), mas colorida.

A ideia é que a função crie um conjunto de **n** imagens, todas com os "quadrados menores" aproximadamente centralizados nos "quadrados maiores". O tamanho dos "quadrados menores" deve ser igual para todas as imagens do conjunto e definido de forma aleatória. As duas cores utilizadas devem variar para cada imagem e também são definidas aleatoriamente.

Você deve definir certos limites para garantir que a geração das imagens faça sentido: qual a altura mínima das imagens? qual o tamanho mínimo dos quadrados menores? qual a distância mínima entre as bordas dos quadrados menor/maior (para que um não se sobreponha ao outro)?

Teste a função com diferentes valores de entrada, tentando sempre plotar as imagens de cada conjunto gerado em um único plot. Comente sobre as decisões tomadas na criação da função (como os limites que foram definidos) e seu comportamento esperado (como a centralização dos quadrados menores em relação às dimensões da imagem).

```python
import random
import numpy as np
import matplotlib.pyplot as plt

def square_generator(h, n):

  array_data = []

  array = np.zeros([h, h*2, 3], dtype=np.uint8)


  len_alt = int(array.shape[0]/2) # metade da altura do quadrado externo
  len_larg = int(array.shape[1]/2) # metade da largura do quadrado externo

  limite_inf = int(len_alt/10) # limite inferior da imagem, serve para garantir que a imagem interna nunca seja
  #igual a zero ou tenha valores muito pequeno
  limite_sup = int((len_alt-10)) # limite superior da imagem, serve para garantir que a imagem interna
  #nunca seja tão grande quanto a externa
  rand_size = random.randint(limite_inf, limite_sup) # gera numeros aleatorios que vai de um décimo da altura da
  #imagem até a metade da altura da imagem menos 10

  largura_max = len_larg + rand_size # a largura maxima do quadrado interno
  largura_min = len_larg - rand_size # a largura miníma do quadrado interno

  altura_max = len_alt + rand_size # a altura máxima do quadrado interno
  altura_min = len_alt - rand_size # a altura minima do quadrado interno
  for i in range(n):
    color1 = np.random.randint(0, 256, size=3) # gerador de cores aleatórios para o quadrado interno
    color2 = np.random.randint(0, 256, size=3) # gerador de cores aleatórios para o quadrado externo

    array[:,:,:] = color2 # Seleciona todas as linhas e colunas e atribui o color2

    array[altura_min:altura_max,largura_min:largura_max:] = color1 # Seleciona as larguras e alturas geradas de forma randomica e atribui o color1
    plt.figure(figsize = (3,3)) # cria uma nova figura de tamanho h por 2*h
    plt.imshow(array,vmin=0, vmax=255, cmap='gray') # mostra a figura
    array_data.append(array)

  return array_data

square_generator(50, 1) #  chama a função com os atributos de altura e numero da imagem a ser gerada
```
